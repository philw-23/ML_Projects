```python
# Base packages
import pandas as pd
import numpy as np
import zipfile
```

```python
# Unzip Data
with zipfile.ZipFile('./raw_data.zip', 'r') as z:
    z.extractall('./')

# Read in data
cc_dat = pd.read_csv('./creditcard.csv')
cc_dat.head()
```

```python
# Summarize Data
cc_dat.info()
```

```python
# Count Null Values
cc_dat.isnull().sum() # In this case expect no nulls, but worth doing for practice
```

```python
# Plot some histograms
import matplotlib.pyplot as plt
if 'Class' in cc_dat.columns: # Prevent error from re-running cell
    cc_dat.drop(columns=['Class'], axis=1, inplace=True) # Drop labels

ax = cc_dat.hist(bins=100)
for axis in ax.flatten(): # Code to remove labels
    axis.set_xticklabels([])
    axis.set_yticklabels([])
```
